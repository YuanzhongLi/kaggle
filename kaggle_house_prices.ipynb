```python
import pandas as pd 
df_train= pd.read_csv('Downloads/train (1).csv')
df_test=pd.read_csv('Downloads/test.csv')
y_train=df_train['SalePrice'].values
df_X_train=df_train.drop('SalePrice',axis=1)
```

```python
#get dummy する際に特徴量に差が出ないようにするためにtrainとtestを一度結合する
df_all=pd.concat([df_X_train, df_test])
df_all.head()
```

```python
#missing data
total= df_all.isnull().sum().sort_values(ascending=False)
percent = (df_all.isnull().sum()/df_all.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(40)
```

#一般的には15%以上が欠損した場合で特徴量ごと消すが、ここではpreprocessingを簡易化するために一つでも欠損値がある特徴量は削除することにする

```python
#欠損値を含む特徴量の行を消す
df_all=df_all.drop((missing_data[missing_data['Total'] >= 1]).index,1)
df_all.head()
```

```python
df_all.dtypes
```

```python
df_all=pd.get_dummies(df_all)
df_all.head()
```

```python
df_X_train=df_all[:1460]
df_X_test=df_all[1460:]
#再びtrainとtestに分ける
```

```python
X_train=df_X_train.values
X_test=df_X_test.values
#DataFrameからNumpyArrayにする
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
X_train_std=sc_x.fit_transform(X_train)
y_train_std=sc_y.fit_transform(y_train)
X_test_std=sc_x.fit_transform(X_test)
```

1.KNNでの回帰

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
print('train set R^2 prediction:{:.2f}'.format(knn.score(X_train,y_train)))
#R^2:coefficient of determination 回帰モデルの予測の正確さを測る指標
#標準化なし
```

```python
knn.fit(X_train_std,y_train_std)
```

```python
print('train_std set R^2 prediction:{:.2f}'.format(knn.score(X_train_std,y_train_std)))
#標準化あり
```

```python
sc_y.inverse_transform(knn.predict(X_test_std))
```

#標準化ありの方がR^2値が高かったのでそっちの方でtest_dataの予測を行った

問題としてはget dummy のところも標準化してしまってる

4.NeuralNetworkでの回帰

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout
from keras.optimizers import Adam
from sklearn.metrics import r2_score
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
from sklearn.metrics import mean_squared_error
```

```python
def reg_model():
    model = Sequential()
    model.add(Dense(output_dim=150, input_dim=150, activation='relu'))
    model.add(Dense(output_dim=150, activation='relu'))
    model.add(Dense(output_dim=150, activation='relu'))
    model.add(Dense(output_dim=1))
    #Fully Connect
    # compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

#入力層のサイズ:150
#中間層の層数:3
#中間層のユニット数:150
#活性化関数:Relu
#loss function:mean_squared_error（平均二乗誤差）
#optimizer:adam(CS231nの第7回参照)

```python
estimator = KerasRegressor(build_fn=reg_model, epochs=100, batch_size=50, verbose=1)
# 標準化なしでトレーニングを行う
estimator.fit(X_train, y_train)
```

#標準化しないとlossがとんでもない値のまま全く減らない

```python
estimator.predict(X_test)
#標準化なしでの予測
```

```python
print('train set R^2 prediction:{:.2f}'.format(r2_score(y_train, estimator.predict(X_train))))
```

#これは何の値かな（汗

```python
estimator = KerasRegressor(build_fn=reg_model, epochs=100, batch_size=50, verbose=1)
# 標準化ありでトレーニングを行う
estimator.fit(X_train_std, y_train_std)
```

#epochは40を超えてからあまり変化がない

```python
sc_y.inverse_transform(estimator.predict(X_test_std))
```

```python
print('train_std set R^2 prediction:{:.2f}'.format(r2_score(y_train_std, estimator.predict(X_train_std))))
```
